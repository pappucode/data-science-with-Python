## Part 1 : Build XGBoost Model

### 1.1 : Read Data

Import Jupyter Notebook in the the current environment that does the Data Processing

```python
"""Import Jupyter Notebook"""

import import_ipynb
import bootcamp_data_processing as bdp
```

Get the pre-processed data

```python
train_df, test_df = bdp.train_test_ordinal
# train_df, test_df = bdp.train_test_onehot
```

```python
# train_df.drop(columns=['Boat'], inplace=True)
# test_df.drop(columns=['Boat'], inplace=True)
```

#### Create preditors (X) and response (y) to be used as model inputs

```python
X = train_df.drop('Survived', axis=1)
y = train_df.Survived
```

### 1.2 : Split Data

Split training dataset in to train and valid datasets

```python
"""Setting the Seed"""
theSeed = 42
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.75, random_state=theSeed)
print(f'train data: {X_train.shape}')
print(f'valid data: {X_valid.shape}')
```

Format test dataset

```python
#X_test = test_df
X_test = test_df.drop('Survived', axis=1)
y_test = test_df.Survived
print(f'test data : {X_test.shape}')
```

### 1.3 : Scale the Data

```python
from sklearn.preprocessing import MinMaxScaler, Normalizer, StandardScaler

scalar  = MinMaxScaler()
X_train = scalar.fit_transform(X_train)
X_valid = scalar.fit_transform(X_valid)
X_test  = scalar.fit_transform(X_test)
```

### 1.4 : Train Model

#### 1.4.1 : Setting Weight to the Data

```python
# scale_pos_weight = np.random.rand(X_train.shape[0], )
```

#### 1.4.2 : XGB Model

XGB Parameter List: https://xgboost.readthedocs.io/en/stable/parameter.html

#### 1.4.3 : Define model

```python
# param_grid = {
#     'objective': 'binary',
#     'booster': 'gbtree', 
#     'tree_method': 'auto',
#     'eval_metric' : ['auc','logloss','aucpr'],
#     'n_jobs': 2,
#     'n_estimators': 100,
#     'max_depth': 5,
#     'learning_rate': 0.1,
#     'gamma': 1.0,
#     'min_child_weight': 0.001, 
#     'max_delta_step': 0.0,
#     'subsample': 1.0, 
#     'colsample_bytree': 1.0, 
#     'colsample_bylevel': 1.0,
#     'colsample_bynode': 1.0,
#     'reg_alpha': 0.0, 
#     'reg_lambda': 0.0,
#     'scale_pos_weight': 1.0,
#     'random_state': None,
#     'verbose': False}

# model = xgb.XGBClassifier()
# model.set_params( **param_grid )
```

```python
model = xgb.XGBClassifier(
    objective='binary:logistic',
    booster='gbtree', 
    tree_method='auto',
    eval_metric=['auc','logloss','aucpr'],
    n_jobs=2,
    n_estimators=100,
    max_depth=5, 
    learning_rate=0.1, 
    gamma=0,
    min_child_weight=0.001, 
    max_delta_step=0.0,
    subsample=1.0, 
    colsample_bytree=1.0, 
    colsample_bylevel=1.0,
    colsample_bynode=1.0,
    reg_alpha=0.0, 
    reg_lambda=0.0,
    scale_pos_weight=1.0,
    #verbose=0,
    random_state=None
)
```

#### 1.4.4 : Fit model

```python
fit = model.fit(
    X_train, y_train,
    sample_weight=None, 
    base_margin=None,
    eval_set=[(X_valid, y_valid)],  
    #eval_metric=['auc','logloss','aucpr'], 
    early_stopping_rounds=None, 
    verbose=False, 
    xgb_model=None, 
    sample_weight_eval_set=None, 
    #base_margin_eval_set=None, 
    feature_weights=None, 
    callbacks=None
)
```

#### 1.4.4 : Model Performance on Test Data

```python
y_pred = model.predict(
    X_test, 
    output_margin=False, 
    ntree_limit=None, 
    validate_features=True, 
    base_margin=None
)
```

```python
y_score = model.predict_proba(
    X_test, 
    ntree_limit=None, 
    validate_features=True, 
    base_margin=None
)
```

## Part 2 : Build XGBoost Model with Optimized Parameters

### 2.1 GridSearch Method

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

```python
param_grid = {
    'n_estimators': [60,80,100],
    'max_depth': [4,5,6],
    'learning_rate': [0.001,0.1,0.3],
    'gamma': [0.0,0.1,0.2],
    'min_child_weight': [0.0005,0.001,0.005],
    'max_delta_step': [0.0],
    'subsample': [0.80,0.9,1.0], 
    'colsample_bytree': [0.8,0.9,1.0], 
    'colsample_bylevel': [1.0],
    'colsample_bynode': [1.0],
    'reg_alpha': [0.0], 
    'reg_lambda': [0.0]
}
```

```python
model = xgb.XGBClassifier(
    objective='binary:logistic',
    booster= 'gbtree', 
    tree_method= 'auto',
    n_jobs= 2,
    scale_pos_weight=1.0,
    random_state=theSeed
)
```

```python
# from sklearn.metrics import SCORERS
# SCORERS.keys()
```

```python
grid_search = GridSearchCV(model, param_grid, scoring='roc_auc', cv=5, refit=True, verbose=0)
grid_fit = grid_search.fit(X_train, y_train)
```

```python
# grid_search = GridSearchCV(model, param_grid, scoring=['roc_auc','neg_log_loss','f1'], cv=5, refit='roc_auc')
# grid_fit = grid_search.fit(X_train, y_train)
```

```python
print("Best Estimator:", grid_search.best_estimator_)
print("Best Index    :", grid_search.best_index_)
print("Best Params   :", grid_search.best_params_)
print("Best Score    :", grid_search.best_score_)
print("Best Classes  :", grid_search.classes_)
```

```python
sorted(grid_search.cv_results_.keys())
```

```python
means = grid_search.cv_results_['mean_test_score']
stds  = grid_search.cv_results_['std_test_score']
for mean, std, params in zip(means, stds, grid_search.cv_results_['params']):
    print("%0.3f (+/-%0.03f) for %r" % (mean, std*2, params))
```

#### 2.1.1 : Best Model Evaluation

```python
"""Get the Best Estimator"""
bestEstimator = grid_search.best_estimator_
```

```python
"""Build the Best Model using the Best Estimator"""
bestEstimator.fit(
    X_train, y_train,  
    eval_set=[(X_valid, y_valid)],
    eval_metric=['auc','logloss','aucpr'], 
    early_stopping_rounds=None,
    verbose=False)
```

```python
auc = list(bestEstimator.evals_result_.values())[0]['auc']
auc = np.array(auc)
print(f"Mean AUC : { round(auc.mean(),4)}")
loss = list(bestEstimator.evals_result_.values())[0]['logloss']
loss = np.array(loss)
print(f"Mean Loss : { round(loss.mean(),4) }")
aucpr = list(bestEstimator.evals_result_.values())[0]['aucpr']
aucpr = np.array(aucpr)
print(f"Mean Loss : { round(aucpr.mean(),4) }")
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(16,4))
ax[0].plot(auc, label='AUC')
ax[1].plot(loss,label='Loss')

ax[0].legend()
ax[1].legend()
plt.show()
```

#### 2.1.2 : Predict

```python
y_pred = grid_search.predict(X_test) 
```

```python
y_score = grid_search.predict_proba(X_test)
```

#### 2.1.3 : Performance Measure

```python
from sklearn import metrics
```

```python
"""Accuracy Score"""
accuracy = metrics.accuracy_score(y_test, y_pred, sample_weight=None, normalize=True)
print( round(accuracy, 4))
```

```python
"""Balanced Accuracy Score"""
balanced_accuracy = metrics.balanced_accuracy_score(
    y_test, y_pred, 
    sample_weight=None, 
    adjusted=False
)

print( round(balanced_accuracy, 4))
```

```python
"""ROC Curve"""
fpr, tpr, thresholds = metrics.roc_curve(y_test, y_score[:,1])
```

```python
"""AUC"""
auc = metrics.auc(fpr, tpr)
print( round(auc, 4))
```

```python
"""Confusion Matrix"""
metrics.confusion_matrix(y_test, y_pred)
```

```python
tn, fp, fn, tp = metrics.confusion_matrix(y_test, y_pred).ravel()
print(f"TN:{tn}, FP:{fp}, FN:{fn}, TP:{tp}")
```

```python
"""F1 Score"""
f1 = metrics.f1_score(
    y_test, y_pred, 
    labels=None, 
    pos_label=1, 
    average='binary', 
    sample_weight=None
)

print( round(f1, 4))
```

```python
"""Log Loss"""
metrics.log_loss(
    y_test, y_score[:, 1], 
    normalize=True, 
    sample_weight=None, 
    labels=[0,1]
)
```

```python
"""Precision, Recall, Fscore, Support"""
metrics.precision_recall_fscore_support(
    y_test, y_pred, 
    labels=[0,1], 
    pos_label=1, 
    average=None, 
    sample_weight=None
)
```

```python
precision = metrics.precision_score(
    y_test, y_pred, 
    labels=[0,1], 
    pos_label=1, 
    average='binary', 
    sample_weight=None
)

print( round(precision, 4))
```

```python
recall = metrics.recall_score(
    y_test, y_pred, 
    labels=[0,1], 
    pos_label=1, 
    average='binary', 
    sample_weight=None
)

print( round(recall, 4))
```

```python
fig, ax = plt.subplots(figsize=(10,6))

metrics.PrecisionRecallDisplay.from_estimator(
    grid_search, X_test, y_test, 
    sample_weight=None, 
    ax=ax
)

plt.show()
```
